```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
titanic_data = sns.load_dataset("titanic")
titanic_data
```

```python
sns.jointplot (x = "age", y= "survived", kind= "hex", data= titanic_data)
plt.show
```

```python
Animals= "Cheetah", "Puma", "Leopard", "Tiger", "Lynx"
Data= [30, 25, 10, 25, 10]
explode = (0, 0, 0, 0.1, 0)

fig1, ax1 = plt.subplots()
ax1.pie(Data, explode = explode, labels=Animals, autopct= "%1.0f%%", 
         shadow=True, startangle=45)
ax1.axis("equal")

plt.show()
```
